Gradiant Descent Theory     

```python
# Import Libraries
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Temp J(w) and its Gradiant
def J(w):
    return w[0]**2 + 2*w[1]**2

def GradOfJ(w):
    gradJ=np.array([2*w[0],4*w[1]])
    return gradJ


# Gradiant Descent Alguritm
def GDescentAlg(grad,w_init,learningRate=0.5,steps=20):
    w_array=np.array(w_init)
    path=[w_array.copy()]
    for step in range(steps):        
        w_grad=GradOfJ(w_array)
        w_new=w_array-learningRate*w_grad
        path.append(w_new.copy())
    
    return np.array(path)



# Initialize w vector and call Gradiant Descent Function

w_init=[4,4]
path_convex=GDescentAlg(GradOfJ,w_init=w_init,learningRate=0.1,steps=50)


# Visualize output
w1Values=np.linspace(-5,5,400)
w0Values=np.linspace(-5,5,400)
w1,w2=np.meshgrid(w0Values,w1Values)

JValues=J([w1,w2])

plt.figure(figsize=(8,6))
countor_levels=np.logspace(-0.5,3,35)
plt.contour(w1,w2, JValues,levels=countor_levels,cmap="jet")
pathX,pathY= path_convex[:,0],path_convex[:,1]
plt.plot(pathX,pathY,marker='o',color='red',label="gradient descent path")
plt.xlabel("w1")
plt.ylabel("w2")
plt.legend()
plt.show()
```
